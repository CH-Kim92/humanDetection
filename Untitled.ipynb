<h2>Real Time Detection</h2>

```python
cap = cv2.VideoCapture(0)
count = 0
check = False 
while cap.isOpened():
    ret, frame = cap.read()
    # Make detection
    results = model(frame)
    df = results.pandas().xyxy[0]
    df["class"] = df["class"].astype(str).astype(int)
    if df.shape[0] > 0 :
        person_df = df[df["class"]==0]
        person_foot = person_df["ymax"].ge(int(360))
        count += int(person_foot.value_counts(True) * person_df.shape[0])
    img = cv2.line(np.squeeze(results.render()), pt1=(0, 360), pt2=(1280, 360), color=(255, 255, 0), thickness=2, lineType=8, shift=0)
    #cv2.imshow('YOLO', np.squeeze(results.render()))
    cv2.imshow('YOLO', img)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
